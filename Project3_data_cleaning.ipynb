```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
pd.options.display.max_rows = 15
pd.options.display.max_columns = 30
```

```python
df = pd.read_csv('/Users/linkexin/Documents/Udacity/project3/movies.csv')
```

```python
df
```

```python
del df['homepage']
del df['cast']
del df['tagline']
del df['overview']
del df['runtime']
del df['budget']
del df['revenue']
```

```python
df
```

```python
split_ser = df['genres'].str.split('|')
```

```python
split_ser
```

```python
new_list = []
for l in split_ser:
    if type(l) != float and len(l) == 5:
        new_list.append(l)
    elif type(l) == float:
        append_list = []
        for i in range(5):
            append_list.append(np.nan)
        new_list.append(append_list)     
    else:
        append_list = l
        for i in range(5 - len(l)):
            append_list.append(np.nan)
        new_list.append(append_list)
```

```python
new_list[:20]
```

```python
genre_df = DataFrame(new_list, columns = ['genre1', 'genre2', 'genre3', 'genre4', 'genre5'])
```

```python
genre_df
```

```python
df = df.join(genre_df)
```

```python
del df['genres']
```
